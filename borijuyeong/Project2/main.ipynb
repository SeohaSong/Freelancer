```python
%matplotlib inline

import pandas as pd
import time

from SVM_classifier import Model


train_dataset = pd.read_pickle("./data/data_batch_1")
train_X = train_dataset["data"]
train_y = train_dataset["labels"]

test_dataset = pd.read_pickle("./data/test_batch")
test_X = test_dataset["data"]
test_y = test_dataset["labels"]


model = Model("",
              learning_rate=0.0001,
              reg_rate=0,
              n_iter=2000,
              n_fold=3,
              batch_size=40)
model.fit(train_X, train_y)


print("Accury for train dataset")
print(sum(model.predict(train_X) == train_y)/len(train_y))
print("Accury for test dataset")
print(sum(model.predict(test_X) == test_y)/len(test_y))
```
